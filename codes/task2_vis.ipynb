```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the data
trip_counts_df = pd.read_csv('output2/Q1/trip_count.csv')

# Sort the DataFrame based on 'trip_count'
trip_counts_df = trip_counts_df.sort_values(by='trip_count')

# After sorting, you can create the 'business_month' label and plot as before
trip_counts_df['business_month'] = trip_counts_df['business'] + '-' + trip_counts_df['month'].astype(str)

plt.figure(figsize=(10, 6))
plt.bar(trip_counts_df['business_month'], trip_counts_df['trip_count'], color='blue')
plt.yscale('log')  # Set the y-axis to logarithmic scale
plt.title('Trip Count per Business and Month')
plt.xlabel('Business-Month')
plt.ylabel('Trip Count (log scale)')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the x-axis labels
plt.show()
```

```python
# Loading the data for platform's profits
profits_df = pd.read_csv('output2/Q2/total_profit.csv')

# Sort the DataFrame based on 'total_profit'
profits_df = profits_df.sort_values(by='total_profit')

# Create the 'business_month' label for plotting
profits_df['business_month'] = profits_df['business'] + '-' + profits_df['month'].astype(str)

# Plotting the platform's profits histogram
plt.figure(figsize=(10, 6))
plt.bar(profits_df['business_month'], profits_df['total_profit'], color='orange')
plt.yscale('log')  # Set the y-axis to logarithmic scale
plt.title('Platform\'s Profits per Business and Month')
plt.xlabel('Business-Month')
plt.ylabel('Platform\'s Profits (log scale)')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the x-axis labels
plt.show()
```

```python
# Loading the data for driver's earnings
earnings_df = pd.read_csv('output2/Q3/total_earnings.csv')

# Sort the DataFrame based on 'total_earnings'
earnings_df = earnings_df.sort_values(by='total_earnings')

# Create the 'business_month' label for plotting
earnings_df['business_month'] = earnings_df['business'] + '-' + earnings_df['month'].astype(str)

# Plotting the driver's earnings histogram
plt.figure(figsize=(10, 6))
plt.bar(earnings_df['business_month'], earnings_df['total_earnings'], color='green')
plt.yscale('log')  # Set the y-axis to logarithmic scale
plt.title('Driver\'s Earnings per Business and Month')
plt.xlabel('Business-Month')
plt.ylabel('Driver\'s Earnings (log scale)')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to make room for the x-axis labels
plt.show()
```
