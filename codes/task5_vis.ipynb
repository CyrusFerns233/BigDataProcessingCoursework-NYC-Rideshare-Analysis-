```python
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV data into a DataFrame
data = pd.read_csv("op/Q1/avgwaitingtime_task5.csv")

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(data['day'], data['average_waiting_time'], color='skyblue')
plt.xlabel('Day')
plt.ylabel('Average Waiting Time (seconds)')
plt.title('Average Waiting Time by Day')
plt.xticks(data['day'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.tight_layout()
plt.show()
```
